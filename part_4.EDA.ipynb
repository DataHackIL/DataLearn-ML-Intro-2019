## Correlation matrix

```python
import matplotlib.pyplot as plt
import seaborn as sns  # nice visualization package
```

Let only use numeric data (although you can check correlation for other types as well).

```python
numeric_acc = acc.select_dtypes('number')
```

And only look at features with at least ~10% non-missing values.

```python
numeric_acc.columns[numeric_acc.isna().sum() < 900]
```

```python
numeric_acc = numeric_acc[numeric_acc.columns[numeric_acc.isna().sum() < 900]]
```

```python
numeric_acc.shape
```

15 features left!

```python
numeric_acc.head(3)
```

```python
corrmat = numeric_acc.corr() 
```

```python
f, ax = plt.subplots(figsize =(11, 9)) 
sns.heatmap(corrmat, ax = ax, cmap ="YlGnBu", linewidths = 0.1) 
```

* Obvoius positive correlations between the # of billed users and (1) the collection in 21 days (probably of fees) (2) being a payed user.
* Obvious positive correlation between min and max team size.
* What about `has_domain` - `is_gmail` - `disable_promotion`?
* is_gmail has some negative correlation with lead_score, our target.
* Amusing negative correlation between account id the company size; perhaps smaller clients have joined later? We won't use account id anway...


Let's look at absolute correlations to let them pop a bit more...

```python
abscorrmat = abs(numeric_acc.corr())
```

```python
f, ax = plt.subplots(figsize =(11, 9)) 
sns.heatmap(abscorrmat, ax = ax, cmap ="YlGnBu", linewidths = 0.1) 
```

It's easier to see now `lead_score` is correlated to a non-negligble degree with several features.

## Correlation network

```python
!pip install networkx
```

```python
import numpy as np
import networkx as nx
from community import best_partition
import matplotlib.pyplot as plt
```

```python
def corr_network(df, threshold=0):
    C = np.corrcoef(df,rowvar=0)
    A = 1*(C>threshold)
    G = nx.Graph(A)
    G = nx.relabel_nodes(G,dict(zip(G.nodes(),df.columns.values)))
    communities = best_partition(G)

    n = len(set(communities.values()))
    print("{} communities detected for {} features!".format(n, len(df.columns)))
    node_coloring = [communities[node]/n for node in G.nodes()]

    nx.pos = nx.fruchterman_reingold_layout(G, dim=2, k=None, pos=None, fixed=None, iterations=5000, weight='weight', scale=1.0, center=None)

    nx.draw_networkx(G, cmap=plt.get_cmap('jet'), with_labels=True, node_color=node_coloring, font_size=16)
    plt.title('Feature correlation network', fontsize=22)
    fig = plt.gcf()
    fig.set_size_inches(18.5, 10.5)
```

```python
corr_network(numeric_acc)
```

```python
numeric_acc.columns
```

```python
sub_numeric_acc = numeric_acc[['account_id', 'paying', 'collection_21_days','disable_promotion',
    'billed_users_count', 'has_domain', 'lead_score']]
```

```python
corr_network(sub_numeric_acc)
```

Let's play push the threshold a bit...

```python
corr_network(sub_numeric_acc, threshold=0.1)
```

And that's probably enough:
